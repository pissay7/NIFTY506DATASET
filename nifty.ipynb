```python

Nifty = pd.read_csv('NIFTY50-18_23.csv', parse_dates = True, index_col='Date' )
```

```python
print(Nifty.head(10))
```

```python
Nifty['Return'] = (Nifty['Close'] - Nifty['Open']) / Nifty['Open']
```

```python
print(Nifty.head(10))
```

```python
ax1 = Nifty['Close'].plot(style = ['-'])
plt.title('Price')
ax1.legend()
plt.plot()
```

```python
Nifty.plot(subplots = True, figsize=(10, 12))
```

```python
ax1 = Nifty['Return'].plot(style = [('-')])
plt.title('Return')
ax1.legend()
plt.plot()
```

```python
Nifty['Close'].hist()
```

```python
Nifty['Return'].hist()
```

```python
#plot kernel density
Nifty['Return'].plot(kind = 'kde')
plt.show()
```
